# Вывод

Вывод - передача данных от компьютера пользователю.\
Для вывода в python используется функция `print()`.\
У функции всегда есть скобки, даже если они пустые. В скобках пишется *аргумент* функции - данные, с которыми функция должна выполнить действие.

```python
print(5) # Печатает число
```

```python
print('hello') # Печатает слово
```

Кавычки в python могут быть одинарными, двойными или тройными. Можно использовать любые, главное - чтобы открывающая и закрывающая кавычки были одинаковыми.

Все, что написано в кавычках, программа прочитывает буквально: каждый символ - это тот символ, который он изображает.

Все, что написано без кавычек, программа прочитывает как руководство к действию: как функцию, оператор или переменную.

```python
print(3, 'hi', 15) # Можно передать несколько аргументов, перечислив их через запятую. Они напечатаются через пробел
```

```python
print(1, 4, 8, 2, sep='; ') # Аргумент sep определяет, какой разделитель будет использован вместо пробела
```

```python
# Если запустить несколько функций print(), каждая выполнится на своей строке,
# потому что по умолчанию выполнение функции print() завершается переносом строки
print(6)
print('bazinga')
print(15)
print(1)
```

```python
# Аргумент end меняет перенос строки в конце выполнения функции print() на ваш символ
print(6, end='! ')
print('bazinga', end='! ')
print(15, end='! ')
print(1, end='! ')
```

# Ввод, переменные

Ввод - передача данных от пользователя компьютеру

```python
input() # Получает данные от пользователя
```

```python
input(5) # 5 в аргументе input() не прочитывается как данные от пользователя
```

```python
input('Введите число ') # Потому что в аргументе input() пишется подсказка пользователю:
                        # чтобы никто не запутался, что нужно ввести
```

Для того, чтобы после получения данных использовать их позже в другом месте, их нужно где-то сохранить и подписать. Для этого используются переменные.

```python
a = input('Введите число') # a - имя переменной; = - оператор присваивания значения переменной
```

```python
print(a)
```

Название переменной может состоять из букв, цифр и нижних подчеркиваний. Переменную можно назвать как угодно, для python конкретное название переменной не имеет значения.

Правильно называть переменные так, чтобы их имена на английском языке отражали, что будет записано в переменной.

```python
my_variable_1 = input('Введите значение переменной ')
```

```python
print(my_variable_1)
```

А если мы напишем my_variable_1 в кавычках, то напечатается my_variable_1, а не значение нашей переменной. О том, как работают кавычки, написано выше.

```python
print('my_variable_1')
```

```python
# Почему так писать неправильно?
print('Введите имя ')
name = input()
```

Этот код неправилен логически и не аккуратен. `print()` должен передавать данные пользователю. "Введите имя" - это не какие-то данные, а подсказка для пользователя. А у подсказки для пользователя есть свое место, оно - в скобках у `input()`. Хотя визуально разницы в выполняющемся коде не будет.

# Строки, числа

Функция `type()` выводит тип данных своего аргумента.

```python
type(5) # int - целое число
```

```python
type('5') # str - строковые данные
```

```python
str(5) # можем сделать число строкой
```

```python
int('9') # а можем сделать строку числом
```

```python
int('привет') # но не любую строку. Сделать число из "привет" сложно
```

```python
int(5.0) # int - целое число. Если привести число с пустой десятичной частью к int,
         # получится просто целое число: без точки и нуля.
```

Строки и числа отличаются тем, что с ними можно делать: строки читают, числа считают. С числами можно выполнять вычисления, со строками это не получится. При этом у строк есть свои возможности, которых нет у чисел.

# Математические операции

```python
3 + 5 # сложение
```

```python
7 - 1 # вычитание
```

```python
6 * 3 # умножение
```

```python
15 / 3 # деление
```

```python
type(15 / 3) # а это - тип данных "число с десятичной частью"
```

При обычном делении мы всегда получаем число с десятичной частью, потому что при делении бывает, что получается дробь.

```python
int(15 / 3) # не хотим точку - после деления делаем число целым
```

```python
7 // 3 # целая часть от деления
```

```python
7 % 3 # остаток от деления
```

```python
3 ** 2 # возведение в степень
```

```python
print(15 / 3) # напечаталось значение выражения
```

```python
print('15 / 3') # напечаталось в точности то, что в кавычках. Вспоминаем, что делают кавычки
```

```python
n = input('Введите число')
print(n * 3)
```

По умолчанию результат получения данных через input() - всегда строка. Строку нельзя математически умножить на число. Поэтому строка при умножении копируется.

```python
n = int(input('Введите число'))
print(n * 3)
```

Чтобы получить тот результат, которого мы ожидали от умножения, при получении данных сразу же приводим их к числовому типу данных. Тогда эти данные будут работать (и умножаться) как число.

# Действия со строками

```python
'само' + 'лёт' # как складываются строки. Эта операция называется конкатенацией
```
