# Словари

Словарь - это тип данных, который позволяет хранить данные с "подписями". Словарь выглядит как набор пар "ключ - значение" и пишется в фигурных скобках. Ключи от значений отделяются двоеточием, между разными парами - запятые. Ключи всегда уникальны, значения могут повторяться. В качестве и ключей, и значений может использоваться любой тип данных, но на практике в качестве ключей удобнее всего использовать строки. Тип данных обозначается dict().

```python
mendeleev_table = {"кислород": "O", "гелий": "He", "дубний": "Db", "золото": "Au"}  # ключ: значение
gradebook = {"Английский": 5, "Физическая культура": 4, "Алгебра": "отлично", "Геометрия": "неуд"}
empty_dict = {}  # пустой словарь
empty_dict2 = dict()  # функция dict() используется для конструирования словарей
```

Порядок значений в словаре неважен, потому что обратиться к любому значению в словаре мы можем по его ключу, написав его после переменной, в которой лежит словарь, в квадратных скобках.

```python
print(mendeleev_table["золото"])
```

Очень похожей записью мы можем добавить в словарь новую пару ключ-значение.

```python
mendeleev_table["висмут"] = "Bi"
print(mendeleev_table)
```

Добавить к словарю пары ключ—значение из другого словаря можно с помощью метода .update():

```python
rare_earth_metals = {"скандий": "Sc", "иттрий": "Y", "лантан": "La"}
mendeleev_table.update(rare_earth_metals)
print(mendeleev_table)
```

Если в добавляемом словаре уже есть пара с тем же ключом, то значение будет изменено на новое. Например:

```python
d1 = {0: "ноль", 1: "один"}
d2 = {1: "единица", 2: "двойка"}
d1.update(d2)
print(d1)
```

Если вам необходимо получить список, состоящий из ключей словаря, используйте метод .keys().

```python
keys = list(mendeleev_table.keys())
print(keys)
```

Для получения списка значений можно использовать метод .values()

```python
keys = list(mendeleev_table.values())
print(keys)
```

Словарь является **неупорядоченной структурой данных**, и поэтому порядок, в котором будут выдаваться ключи или значения словаря, может изменяться от запуска к запуску программы, при использовании другой версии Python или при изменении погоды на Марсе. **Никогда не полагайтесь на то, что порядок значений словаря не изменится при следующем запуске программы.** Таким образом, вы всегда должны считать, что все списки, создаваемые с помощью методов .keys() и .values() имеют случайный порядок.
