* По какому принципу работает if? Как он определяет, когда выполнять одну часть кода, а когда - другую?
* Что, помимо неравенства, мы можем написать в качестве условия в if?

```python
if True:
  print(1)
else:
  print(2)
```

```python
if False:
  print(1)
else:
  print(2)
```

```python
# Новый логический оператор not
if not False:
  print('ЁЖИК')
else:
  print('ОЛЕНЬ')
```

```python
name = input('Введите имя ')

if not name:
  print('Вы ничего не ввели')
else:
  print('Ваше имя ' + name)
```

```python
name = input('Введите имя ')

if name == '':
  print('Вы ничего не ввели')
else:
  print('Ваше имя ' + name)
```
