# Цикл while

Циклов в Python два вида. Первый - while. Разберем на примере.

```python
a = 0

while a < 5:
  print(a)
  a += 1 # то же самое, что a = a + 1
```

После слова `while` мы написали условие выполнения цикла. Вспоминаем, что такое условия, и понимаем, что условия могут быть либо истинны, либо ложны.

Кроме того, обращаем внимание на то, что в коде есть две строки, *вложенные* в наш while, то есть написанные с отступом. Этот вложенный код называется *телом цикла*. Тело цикла выполняется только тогда, когда условие цикла истинно. Как только условие станет ложным, выполнение тела цикла прекратится.

Теперь смотрим на первую строку. Мы задали начальное значение переменной, которую будем проверять в условии цикла. Если значение этой переменной не будет меняться, условие цикла никогда не станет ложным. Значит тело цикла будет выполняться бесконечно. Поэтому на каждом повторении мы меняем значение переменной, увеличивая его на один. Конструкция `+= 1` увеличивает значение переменной на 1 и называется *инкрементом*. Точно так же можно использовать *декремент* `-= 1` или менять значение инкремента и декремента: изменять значение переменной не на 1, а на любое другое число. Кстати, с умножением и делением такая конструкция тоже работает.
