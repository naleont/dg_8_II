# Очистка данных

## `.split()`, `.strip()`

Есть еще один метод строк, который мы еще не обсуждали. Метод `.strip()` удаляет из начала и из конца строки все символы, переданные в его аргумент. Если аргумент не передан, метод будет работать со всеми пробельными символами.

```python
line = ' hello       '
line.strip()
```

```python
line = '!!!!!hel!lo!!'
line.strip('!')
```

```python
line = ' ?36!.,hello6'
line.strip('? !.,63')
```

Этот метод незаменим для очистки данных. Он часто применяется вместе с `.split()`.

```python
line_list = '1, 2,3,4, 5, 6,  7  , 8,9'
line_list.split(', ')
```

```python
line_list.split(',')
```

```python
list_clean = []
for element in line_list.split(','):
    list_clean.append(element.strip())

print(list_clean)
```

## Работа с регистром

Бывает необходимость распознавать простые ответы "да" и "нет", при этом пользователь может написать "Да", "ДА", "дА", "да" - и с точки зрения компьютера это все разные слова. Для того, чтобы распознавать эти варианты как один, удобно привести ответ пользователя к нижнему регистру, для этого полезен метод строк `.lower()`.

```python
word = 'пОпакАТЕпЕтЛЬ'
word.lower()
```
